```python
import sys
import os

import pandas as pd




module_path = os.path.abspath(os.path.join(".."))
if module_path not in sys.path:
    sys.path.append(module_path)
from irms import data_wrangling

from irms.data_analysis import label_peaks, data_analysis
from irms.constants import *

DEFAULT_DATA_PATH = (
    "~/Desktop/lab_work/orphan/isotope_ratio_mass_spec/data/shay-2022-12-02"
)
DEFAULT_FILE_NAME = "EN_HBG_data_batch2.xlsx"


def _label_peaks_first_pass(
    data_path=DEFAULT_DATA_PATH, filename=DEFAULT_FILE_NAME
) -> pd.DataFrame:
    input_file_name = os.path.join(data_path, filename)
    df = data_analysis.label_aa_peaks_based_on_manual_analysis(input_file_name)
    return df

def _label_peaks(data_path=DEFAULT_DATA_PATH, filename=DEFAULT_FILE_NAME, labeled_df=None
) -> pd.DataFrame:
    input_file_name = os.path.join(data_path, filename)
    df = data_analysis.label_aa_peaks_based_on_avg_time_to_ALA(input_file_name, labeled_df=labeled_df)
    return df



def read_peaks(input_file_name) -> pd.DataFrame:
    df = pd.read_csv(input_file_name)
    df.head(10)
    return df


df = _label_peaks_first_pass()
df = _label_peaks(labeled_df=df)

# df = read_peaks("labeled_en_hbg_data.csv")
df.head(10)
```

```python
from irms.constants import name_to_short_name

def get_aa_mean_relative_time(df):
    aa_peak_summary_data = {}
    for group, data in df.groupby("TENTATIVE_COMPOUND"):
        short_name = name_to_short_name[group]
        aa_peak_summary_data[short_name] = {}
        aa_peak_summary_data[short_name]['mean_alanine_relative_time'] = data["TIME_RELATIVE_TO_ALANINE"].mean().round()
    aa_peak_summary_data.pop("H2")
    aa_peak_summary_data.pop("ALA")
    return aa_peak_summary_data




def get_time_variances(aa_peak_summaries, df):
    for amino_acid, aa_data in aa_peak_summaries.items():
        time_variance = data_analysis.find_time_variance(df, amino_acid, aa_data['mean_alanine_relative_time'])
        aa_peak_summaries[amino_acid]['time_variance'] = time_variance
    return aa_peak_summaries

path_to_data = os.path.join(DEFAULT_DATA_PATH, DEFAULT_FILE_NAME)
original_df = data_wrangling.load_data_frame_from_excel_file(path_to_data)
clean_df = data_wrangling.clean_up_data(original_df)
clean_df = label_peaks.label_H2_and_alanine(clean_df)
aa_mean_relative_time = get_aa_mean_relative_time(df)
aa_peak_summaries = get_time_variances(aa_mean_relative_time, clean_df)
```
