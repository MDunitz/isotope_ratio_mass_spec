```python
import sys
import os
import numpy as np

module_path = os.path.abspath(os.path.join(".."))
if module_path not in sys.path:
    sys.path.append(module_path)
from irms import data_wrangling
from irms.data_analysis import label_peaks
from irms.constants import *

data_path = "~/Desktop/lab_work/orphan/isotope_ratio_mass_spec/data/shay-2022-12-02"
filename = "EN_HBG_data_batch2.xlsx"
input_file_name = os.path.join(data_path, filename)
original_df = data_wrangling.load_data_frame_from_excel_file(input_file_name)
df = data_wrangling.clean_up_data(original_df)
df = label_peaks.label_H2_and_alanine(df)
df = label_peaks.label_required_aas(df)
df = label_peaks.label_non_required_aas(df)
df.head(20)
```

```python
# Methionine"
# MET:0.9,  SER: .75, LYS: 0.67, TYR: 0.67}
TIME_VARIANCE = 5

df = df.groupby("Analysis", group_keys=False).apply(
    label_peaks.create_time_relative_to_glutamic_acid
)
df = df.groupby("Analysis", group_keys=False).apply(
    label_peaks.create_time_relative_to_phenylalanine
)
source_of_truth = df[df["Analysis"] == 4167][
    [
        "TIME_RELATIVE_TO_ALANINE",
        "TIME_RELATIVE_TO_GLUTAMIC_ACID",
        "TIME_RELATIVE_TO_PHENYLALANINE",
        "peak_order",
        "compound",
        "amplitude_2",
        "area_2",
        "bio_replicate_id",
        "TENTATIVE_COMPOUND",
        "color",
    ]
]
methionine_time = int(
    source_of_truth[source_of_truth["compound"] == "Met"][
        "TIME_RELATIVE_TO_ALANINE"
    ].values[0]
)
methionine_time_to_GLU = int(
    source_of_truth[source_of_truth["compound"] == "Met"][
        "TIME_RELATIVE_TO_GLUTAMIC_ACID"
    ].values[0]
)
methionine_time_to_PHE = int(
    source_of_truth[source_of_truth["compound"] == "Met"][
        "TIME_RELATIVE_TO_PHENYLALANINE"
    ].values[0]
)


possible_methionine_by_alanine = df[
    (df["TIME_RELATIVE_TO_ALANINE"] <= methionine_time + TIME_VARIANCE)
    & (df["TIME_RELATIVE_TO_ALANINE"] >= methionine_time - TIME_VARIANCE)
]
possible_methionine_by_glu = df[
    (df["TIME_RELATIVE_TO_GLUTAMIC_ACID"] <= methionine_time_to_GLU + TIME_VARIANCE)
    & (df["TIME_RELATIVE_TO_GLUTAMIC_ACID"] >= methionine_time_to_GLU - TIME_VARIANCE)
]
possible_methionine_by_phe = df[
    (df["TIME_RELATIVE_TO_PHENYLALANINE"] <= methionine_time_to_PHE + TIME_VARIANCE)
    & (df["TIME_RELATIVE_TO_PHENYLALANINE"] >= methionine_time_to_PHE - TIME_VARIANCE)
]

df.loc[
    possible_methionine_by_alanine.index, ["TENTATIVE_COMPOUND_BASED_ON_ALANINE"]
] = amino_acids["MET"]["full_name"]
df.loc[
    possible_methionine_by_glu.index, ["TENTATIVE_COMPOUND_BASED_ON_GLU"]
] = amino_acids["MET"]["full_name"]
df.loc[
    possible_methionine_by_phe.index, ["TENTATIVE_COMPOUND_BASED_ON_PHE"]
] = amino_acids["MET"]["full_name"]

ala_set = set(possible_methionine_by_alanine.index)
phe_set = set(possible_methionine_by_phe.index)
glu_set = set(possible_methionine_by_glu.index)

print(f"total sample count: {len(df['sample_id'].unique())}")
print(f"time variance: {TIME_VARIANCE}")
print(
    f"met by ala: {len(possible_methionine_by_alanine['sample_id'].values)}, unique sample ids: {len(possible_methionine_by_alanine['sample_id'].unique())}"
)
print(
    f"met by glu: {len(possible_methionine_by_glu['sample_id'].values)}, unique sample ids: {len(possible_methionine_by_glu['sample_id'].unique())}"
)
print(
    f"met by phe: {len(possible_methionine_by_phe['sample_id'].values)}, unique sample ids: {len(possible_methionine_by_phe['sample_id'].unique())}"
)

print(
    f"index intersection len: {len(ala_set.intersection(phe_set).intersection(glu_set))}"
)
print(f"index union len: {len(ala_set.union(phe_set).union(glu_set))}")
print(f"ala mean: \n{possible_methionine_by_alanine[['amplitude_2', 'area_2']].mean()}")
print(f"glu mean: \n{possible_methionine_by_glu[['amplitude_2', 'area_2']].mean()}")
print(f"phe mean: \n{possible_methionine_by_phe[['amplitude_2', 'area_2']].mean()}")
potential_indexes = np.array(list(ala_set.union(phe_set).union(glu_set)))
hold = df.loc[potential_indexes]
len(hold["sample_id"].unique())
```
